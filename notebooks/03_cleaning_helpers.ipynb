```python
import os
import pandas as pd

# Paths

CSV_PATH = "/Users/kristofferneo/Desktop/csc173_deepcomputer_vision/data/processed/person_clean_dataset.csv"
IMAGE_DIR = "/Users/kristofferneo/Desktop/csc173_deepcomputer_vision/data/raw_images/train/person"
OUT_CSV  = "/Users/kristofferneo/Desktop/csc173_deepcomputer_vision/data/processed/person_clean_dataset_filtered.csv"

# Load CSV
df = pd.read_csv(CSV_PATH)

# Collect valid image IDs on disk
valid_image_ids = {
    os.path.splitext(fname)[0]
    for fname in os.listdir(IMAGE_DIR)
    if fname.lower().endswith(".jpg")
}

print(f"[INFO] Found {len(valid_image_ids)} image files on disk")

# Filter rows whose image_id exists
mask = df["image_id"].astype(str).isin(valid_image_ids)
df_clean = df[mask].reset_index(drop=True)

# Report stats
removed = len(df) - len(df_clean)
print(f"[INFO] Removed {removed} rows with missing images")
print(f"[INFO] Remaining rows: {len(df_clean)}")

# Save cleaned CSV
df_clean.to_csv(OUT_CSV, index=False)
print(f"[INFO] Saved cleaned dataset to:\n{OUT_CSV}")
```
